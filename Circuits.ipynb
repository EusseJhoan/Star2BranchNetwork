In this notebook, we create the four circuits needed to reproduce the four possible measurement settings of Alice $A_i$ and Charlie $C_j$. We use the library qiskit from IBM to implement it.\
It is important to highlight that in this notebook we are not running the circuits, we ran them directly on the IBM Quantum Plataform.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
```

$A_0=C_0=\frac{\sigma_x +\sigma_y}{\sqrt{2}}$ 

```python
qreg_q = QuantumRegister(5, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

#prepare the maximally entangled states |phi+> between qubits 1 and 2, and qubits 4 and 3
circuit.h(qreg_q[4])
circuit.h(qreg_q[1])
circuit.cx(qreg_q[1], qreg_q[2])
circuit.cx(qreg_q[4], qreg_q[3])
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4])

#perform Alice and Charlie measurements
circuit.t(qreg_q[4])
circuit.h(qreg_q[4])
circuit.t(qreg_q[1])
circuit.h(qreg_q[1])
circuit.measure(qreg_q[4], creg_c[3])
circuit.measure(qreg_q[1], creg_c[0])
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4])

#perform Bob Bell measurement 
circuit.cx(qreg_q[2], qreg_q[3])
circuit.h(qreg_q[2])
circuit.measure(qreg_q[3], creg_c[2])
circuit.measure(qreg_q[2], creg_c[1])
circuit.draw(output="mpl")
```
